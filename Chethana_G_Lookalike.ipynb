```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
products = pd.read_csv(r'C:\Users\Chethana\Downloads\Products.csv')
customers = pd.read_csv(r'C:\Users\Chethana\Downloads\Customers.csv')
transactions = pd.read_csv(r'C:\Users\Chethana\Downloads\Transactions.csv')
```

```python
# Display the first rows of each dataset
products_info = products.head()
customers_info = customers.head()
transactions_info = transactions.head()
```

```python
# Merge datasets for comprehensive analysis
merged_data = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')
merged_data
```

```python
# Filter for the first 20 customers (C0001 to C0020)
customer_ids = [f'C{i:04d}' for i in range(1, 21)]
filtered_data = merged_data[merged_data['CustomerID'].isin(customer_ids)]
filtered_data
```

```python
#Create Customer Profiles
# Create a pivot table with CustomerID and ProductID counts
customer_profile = filtered_data.pivot_table(index='CustomerID', columns='ProductID', values='Quantity', aggfunc='sum', fill_value=0)
customer_profile 
```

```python
# Add additional features like total purchases and total spent
customer_profile['TotalPurchases'] = filtered_data.groupby('CustomerID')['Quantity'].sum()
customer_profile['TotalSpent'] = filtered_data.groupby('CustomerID')['TotalValue'].sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
scaler = StandardScaler()
normalized_data = scaler.fit_transform(customer_profile)
similarity_matrix = cosine_similarity(normalized_data)
similarity_matrix 
```

```python
lookalikes = {}
for idx, customer_id in enumerate(customer_profile.index):
    similarity_scores = list(enumerate(similarity_matrix[idx]))
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:4]
    lookalikes[customer_id] = [(customer_profile.index[i], score) for i, score in sorted_scores]


lookalike_df = pd.DataFrame.from_dict(
    lookalikes,
    orient='index',
).apply(lambda row: [val for pair in row if pair for val in pair], axis=1)


columns = []
for i in range(1, 4):
    columns.extend([f'Lookalike_{i}', f'Score_{i}'])
lookalike_df = lookalike_df.apply(pd.Series)
lookalike_df.columns = columns[:lookalike_df.shape[1]]

print(lookalike_df)
```

```python
lookalike_output = lookalike_df.reset_index()
lookalike_output
```

```python
lookalike_output.columns = ['CustomerID', 'Lookalike_1', 'Score_1', 'Lookalike_2', 'Score_2', 'Lookalike_3', 'Score_3']
lookalike_output.columns
```

```python
lookalike_output.to_csv('Lookalike.csv', index=False)

print("Lookalike recommendations saved to Lookalike.csv")
```
